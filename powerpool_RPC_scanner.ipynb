```python
DataFrame = pd.DataFrame()
for i in keeper_list: #range(1, 45): #
    #url = "https://app.powerpool.finance/#/sepolia/explorer/keepers/0xbdE2Aed54521000DC033B67FB522034e0F93A7e5/" + str(f"{i}") + "/"
    url = "https://app.powerpool.finance/#/gnosis/explorer/keepers/0x071412e301C2087A4DAA055CF4aFa2683cE1e499/" + str(f"{i}") + "/"
    #url = "https://app.powerpool.finance/#/arbitrum/explorer/keepers/0xad1e507f8A0cB1B91421F3bb86BBE29f001CbcC6/" + str(f"{i}") + "/"
    #url = "https://app.powerpool.finance/#/base/explorer/keepers/0x12e49CeDc34C4F455e0dfff7ec38cC535Cbd07C2/" + str(f"{i}") + "/"
    driver.get(url)
    sleep(5)

    #собираем данные
    html_page = driver.page_source
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    data = [i]
    values = soup.find_all('div', {"class":"value"})
    for v in values[-7:]:
        data.append(v.get_text())
    if len(data)<7:
        for q in range(7):
            data.append('-')

    #заносим в датафрейм
    df = pd.DataFrame(data = [data], columns = header)
    DataFrame = pd.concat([DataFrame, df], ignore_index=True, sort=False)
```
